## Code

Run each cell in order:


### Step 1: BERT input format

Single sentence:  [CLS] The cat sat on the mat [SEP]
Sentence pair:    [CLS] How old are you? [SEP] I am 25 [SEP]


```python

# Using HuggingFace Transformers
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

### Step 2: Tokenize


```python
inputs = tokenizer("The bank by the river", return_tensors="pt")
print(inputs['input_ids'])     # token IDs including [CLS] and [SEP]
print(inputs['token_type_ids'])  # segment IDs (all 0 for single sentence)
print(inputs['attention_mask'])  # 1 for real tokens, 0 for padding
```

### Step 3: Get contextual embeddings


```python
outputs = model(**inputs)
last_hidden = outputs.last_hidden_state  # (1, seq_len, 768)
cls_vector = last_hidden[:, 0, :]         # [CLS] token for classification
print(f"Output shape: {last_hidden.shape}")
print(f"[CLS] vector shape: {cls_vector.shape}")
```
