## Context

**Instruction tuning** transforms a base model into an assistant that follows instructions. The idea, pioneered by FLAN (Google, 2022) and InstructGPT (OpenAI, 2022), is to fine-tune the model on (instruction, response) pairs:

## Key Takeaway

The transformation is dramatic. Before instruction tuning, the model is a creative but aimless text generator. After, it's a helpful assistant that stays on topic, follows formatting requests, admits when it doesn't know, and avoids harmful outputs.
