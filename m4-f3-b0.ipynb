## Context

One critical detail: **dropout in deep RNNs**. You apply dropout *between* layers (on the vertical connections), NOT within a layer across time steps. Dropping connections across time would destroy the temporal information flow — exactly what we've been working so hard to preserve.

> **Note:** In CNNs, early layers detect edges, middle layers detect textures, and deep layers detect objects. What might the layers of a deep RNN capture? Think about language: what are the 'edges' and 'objects' of text?

## Code

Run each cell in order:


### Step 1: Stacked BiLSTM with dropout between layers


```python
import torch
import torch.nn as nn

model = nn.LSTM(
    input_size=300,      # word embedding dimension
    hidden_size=256,     # hidden state size per direction
    num_layers=3,        # 3 stacked LSTM layers
    batch_first=True,
    bidirectional=True,  # BiLSTM → output dim = 2 × 256 = 512
    dropout=0.3,         # dropout between layers (not within time)
)

# Input: batch of 32 sentences, max length 50, embedding dim 300
x = torch.randn(32, 50, 300)
```

### Step 2: Forward pass


```python
output, (h_n, c_n) = model(x)
print(f"Output shape: {output.shape}")   # [32, 50, 512]
print(f"h_n shape:    {h_n.shape}")      # [6, 32, 256] — 3 layers × 2 directions
print(f"c_n shape:    {c_n.shape}")      # [6, 32, 256]

# Total parameters
total = sum(p.numel() for p in model.parameters())
print(f"\nTotal parameters: {total:,}")
```

### Step 3: For comparison: single-layer unidirectional


```python
simple = nn.LSTM(300, 256, num_layers=1, batch_first=True)
simple_params = sum(p.numel() for p in simple.parameters())
print(f"Single-layer params: {simple_params:,}")
print(f"Deep BiLSTM is {total/simple_params:.1f}× larger")
```
