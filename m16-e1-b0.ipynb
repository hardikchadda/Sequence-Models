## Key Takeaway

**SentencePiece** (Google, 2018) extends BPE to be language-agnostic — it works directly on raw bytes, no pre-tokenization needed. This is critical for multilingual models that must handle Chinese, Arabic, Hindi, and code in the same vocabulary.
