## Context

In PyTorch, gradient clipping is a single line: `torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)`. It's applied after `loss.backward()` and before `optimizer.step()`.

> **Note:** Gradient clipping solves exploding gradients but does **nothing** for vanishing gradients. Making a tiny gradient bigger would change its direction meaninglessly. The fix for vanishing gradients requires a structural change to the architecture — which is exactly what LSTM and GRU provide in Module 4
