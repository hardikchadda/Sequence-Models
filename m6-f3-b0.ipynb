```python
import torch
import torch.nn as nn
import numpy as np
```

### Step 2: Loading GloVe embeddings in PyTorch


```python
def load_glove(path, word_to_idx, embedding_dim=300):
    """Load GloVe vectors for words in our vocabulary."""
    embeddings = np.random.randn(len(word_to_idx), embedding_dim) * 0.01
    found = 0
    
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.split()
            word = parts[0]
            if word in word_to_idx:
                idx = word_to_idx[word]
                embeddings[idx] = np.array(parts[1:], dtype=float)
                found += 1
    
    print(f'Found {found}/{len(word_to_idx)} words in GloVe')
    return torch.tensor(embeddings, dtype=torch.float32)
```

### Step 3: Using pretrained embeddings in a model


```python
class TextClassifier(nn.Module):
    def __init__(self, pretrained_embeddings, num_classes, freeze=True):
        super().__init__()
        vocab_size, embed_dim = pretrained_embeddings.shape
        
        # Initialize embedding layer with pretrained vectors
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.embedding.weight = nn.Parameter(pretrained_embeddings)
        
        if freeze:
            self.embedding.weight.requires_grad = False  # Don't update during training
        
        self.fc = nn.Linear(embed_dim, num_classes)
    
    def forward(self, x):
        # x: (batch, seq_len) of word indices
        emb = self.embedding(x)          # (batch, seq_len, embed_dim)
        avg = emb.mean(dim=1)            # (batch, embed_dim) — average pooling
        return self.fc(avg)              # (batch, num_classes)

# Example usage
```

## Key Takeaway

Pretrained embeddings give you a massive head start. A model that would take weeks to train from scratch can reach comparable performance in hours using pretrained embeddings — because the embeddings already encode general language knowledge.
